```python
# ================== Robust loader + lead/lag heatmaps ==================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1) Load robustly and get 'date' ---
path = "/content/master_with_IGREA_GSCPI_CPUI.csv"

def load_with_date(path):
    # 1) BOM까지 처리
    try:
        df = pd.read_csv(path, encoding="utf-8-sig")
    except:
        df = pd.read_csv(path)
    df.columns = [c.strip() for c in df.columns]

    # 2) 'date'가 있으면 끝
    if "date" in df.columns:
        df["date"] = pd.to_datetime(df["date"], errors="coerce")
        return df

    # 3) 첫 컬럼을 날짜로 시도
    try:
        df_try = pd.read_csv(path, parse_dates=[0], encoding="utf-8-sig")
    except:
        df_try = pd.read_csv(path, parse_dates=[0])
    first = df_try.columns[0]
    if np.issubdtype(df_try[first].dtype, np.datetime64):
        df_try = df_try.rename(columns={first: "date"})
        return df_try

    # 4) 칼럼명 중 date 비슷한 거 찾기(BOM/깨짐 방지)
    for c in df.columns:
        if "date" in c.lower():  # e.g., '﻿date', '癤풼ate'
            df = df.rename(columns={c: "date"})
            df["date"] = pd.to_datetime(df["date"], errors="coerce")
            return df

    # 5) 정말 없으면 2015-01부터 월별 생성(행수가 월단위라고 가정)
    start = pd.Timestamp("2015-01-01")
    df["date"] = pd.date_range(start=start, periods=len(df), freq="MS")
    return df

df = load_with_date(path)
df = df.dropna(subset=["date"]).sort_values("date").reset_index(drop=True)
df = df.set_index("date")

# 숫자형으로 강제 변환
for c in df.columns:
    if not pd.api.types.is_numeric_dtype(df[c]):
        df[c] = pd.to_numeric(df[c].astype(str).str.replace(",", ""), errors="coerce")

print("Loaded columns:", list(df.columns)[:8], "... (total:", len(df.columns), ")")

# --- 2) Targets (grain import WEIGHTS) & candidate macros ---
targets = [
    "Wheat_import_weight_kg",
    "Corn_import_weight_kg",
    "Rice_import_weight_kg",
    "OtherGrains_import_weight_kg",
]
candidates = ["WTI_Oil","Brent_Oil","krw_per_usd","USD_Index","IGREA","GSCPI","CPUI"]

# 실제 존재하는 것만 사용
targets = [t for t in targets if t in df.columns]
candidates = [x for x in candidates if x in df.columns]
print("Targets in file:", targets)
print("Candidates in file:", candidates)

# --- 3) Lead/Lag correlation (±6M) ---
max_lag = 6
def lead_lag_table(y, df_vars, xs, max_lag=6):
    res = pd.DataFrame(index=range(-max_lag, max_lag+1), columns=xs, dtype=float)
    for x in xs:
        s = df_vars[x]
        for lag in range(-max_lag, max_lag+1):
            if lag > 0:   # x leads
                xc, yc = s.shift(lag), y
            elif lag < 0: # y leads
                xc, yc = s, y.shift(-lag)
            else:
                xc, yc = s, y
            res.loc[lag, x] = xc.corr(yc)
    return res

for tgt in targets:
    tbl = lead_lag_table(df[tgt], df, candidates, max_lag=max_lag)

    plt.figure(figsize=(12,7))
    sns.heatmap(tbl.T, annot=True, fmt=".2f", cmap="coolwarm", center=0)
    plt.title(f"Lead/Lag Correlation with {tgt} (±{max_lag}M)")
    plt.xlabel("Lag (months)   →   Positive = variable leads")
    plt.ylabel("Variables")
    plt.tight_layout()
    plt.show()

    # Top-5 라그 요약
    best = tbl.abs().stack().sort_values(ascending=False).head(5)
    print(f"[Top-5 lags | {tgt}]")
    for (lag, var), _ in best.items():
        direction = "var leads" if lag > 0 else "target leads" if lag < 0 else "coincident"
        print(f"  {var:>10s} @ lag {lag:+}M  corr={tbl.loc[lag,var]:+.3f}  ({direction})")
    print()

```

###📌 **곡물별 민감도 (Elasticity) 분석:분석 개요**

**목표**:
곡물 수입량(kg) 또는 수입금액(USD)이

유가(WTI, Brent),

환율(krw_per_usd),

달러인덱스(USD_Index),

에 얼마나 민감하게 반응하는지 정량적으로 추정.

**방법:**

각 곡물별로 수입량 로그 변환 (ln Q).

설명 변수들도 로그 변환 (ln X).

결과 β 값이 바로 탄력성(Elasticity) → “변수가 1% 변할 때 곡물 수입량이 몇 % 변하는가”.


###**-민감도_회귀(탄력성) 분석 결과 해석**
##1. 밀 (Wheat)

R² = 0.22 → 설명력이 낮음 (거의 설명 안 됨).

모든 변수의 계수 유의성이 p-value > 0.1 → 통계적으로 의미 있는 탄력성 없음.

→ 밀 수입량은 유가·환율 등 주요 거시 변수와의 직접적 선형 관계가 약하다는 결론. (단기 변동성·정책적 요인이 더 중요할 수 있음)

##2. 옥수수 (Corn)

R² = 0.21 → 설명력이 낮음.

마찬가지로 유의한 변수 없음.

→ 옥수수는 상관분석에서는 환율/달러 지수에 민감했지만, 회귀(동시 모형) 에서는 잡히지 않음 → 다중공선성(조건수 3.16e+03) 문제 가능성.

즉, 환율·달러지수·유가가 동시에 들어가면 신호가 상쇄됨.

##3. 쌀 (Rice) ⭐

R² = 0.54 → 설명력이 높음 (곡물 중 가장 잘 설명).

유의한 변수:

WTI 유가 (+46.7, p=0.027) → 유가가 오르면 쌀 수입량 증가.

Brent 유가 (–45.7, p=0.022) → Brent는 반대로 감소. (WTI·Brent 간 multicollinearity 가능성)

IGREA (–0.74, p=0.013) → 세계경제활동↑ → 쌀 수입량 ↓ (내수·보호무역적 요인일 수 있음)

GSCPI (+1.07, p=0.020) → 공급망 압박↑ → 쌀 수입량↑ (수급 불안 대비 선제 수입일 가능성)

→ 쌀은 세계 경기·공급망 변수에 민감, 정책적 비축 수요와 연결 가능.

##4. 기타 곡물 (OtherGrains)

R² = 0.25 → 설명력 낮음.

모든 변수 유의하지 않음.

→ 기타 곡물은 불규칙적 수입·정책적 요인이 커서 단순 회귀로 설명이 어렵다.

##🎯 종합 인사이트

쌀이 가장 거시 변수와 설명력이 높음

IGREA(–), GSCPI(+), 유가(±) → 국제 경기·공급망 상황에 따라 선제적으로 수입량이 크게 바뀜.

→ 쌀은 중장기 정책적 변수를 반영하는 곡물.

밀·옥수수·기타곡물은 회귀 모형 설명력 낮음

상관분석에서는 민감성이 있었으나, 회귀에서는 다중공선성 때문에 계수가 통계적으로 유의하지 않음.

→ 이 경우, 단변량 회귀(예: 옥수수 vs 환율만), 혹은 주성분분석(PCA)·LASSO 같은 변수 선택 기법으로 다시 분석 필요.

실무적으로

쌀 → 정책적 모니터링 지표 (공급망 압박, 글로벌 경기 동향)

옥수수 → 환율/달러인덱스에 민감 (단일변수 모델로 재확인 필요)

밀/기타 → 단기 충격 요인(무역정책, 국내 생산량, 특정 사건)으로 설명되는 가능성 큼

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

# 데이터 불러오기
path = "/content/master_with_IGREA_GSCPI_CPUI.csv"
df = pd.read_csv(path, encoding="utf-8-sig")


# 날짜 처리
df['date'] = pd.to_datetime(df['date'])
df = df.set_index('date')

# 분석 대상 (곡물별 수입량)
targets = [
    "Wheat_import_weight_kg",
    "Corn_import_weight_kg",
    "Rice_import_weight_kg",
    "OtherGrains_import_weight_kg"
]

# 설명 변수 후보
exog_vars = ["WTI_Oil", "Brent_Oil", "krw_per_usd",
             "USD_Index", "IGREA", "GSCPI", "CPUI"]

# 로그 변환 (음수/0 값은 제거)
df_log = df.copy()
for col in targets + exog_vars:
    df_log[col] = pd.to_numeric(df_log[col], errors="coerce")
    df_log[col] = df_log[col].replace(0, np.nan)
    df_log[col] = np.log(df_log[col])

results = {}

for target in targets:
    y = df_log[target].dropna()
    X = df_log[exog_vars].loc[y.index].dropna()
    y = y.loc[X.index]

    X = sm.add_constant(X)
    model = sm.OLS(y, X).fit()
    results[target] = model

    print(f"\n===== {target} Elasticity Results =====")
    print(model.summary())
```

👉 종합: 옥수수는 거의 모든 변수에서 유의한 양(+)의 영향. 즉, 어떤 거시 충격이 와도 수입량을 줄이지 않고 오히려 늘리는 구조.
➡ “사료 곡물은 수요 비탄력적”이라는 특징이 그대로 반영.

WTI, Brent (+, coef ≈ 1.0~1.3e6, p < 0.005, R²≈0.07~0.09)
→ 유가가 오를수록 밀 수입량 증가. (물류비용↑에도 불구, 가공식품 수요 충족 위해 수입 유지/확대)

환율 (+, coef ≈ 1.6e5, p=0.012)
→ 원/달러 상승 → 밀 수입량 증가. 옥수수와 유사하게 수요 비탄력성 반영.

👉 종합: 밀은 에너지·환율 충격에도 수입 유지/확대되는 구조. 다만 옥수수보다 변수 민감성은 낮음.

환율(krw_per_usd) (–, coef ≈ –1.25e3, p=0.036, R²≈0.035)
→ 원화 약세 → 수입량 감소.

👉 종합: 기타 곡물은 규모가 작아 거시 변수에 민감하게 반응. 환율이 불리할 때 수입이 줄어드는 전형적 반응.

4. 쌀 (Rice)

→ 쌀은 WTO 협정상 수입쿼터·정책비중이 커서, 거시 변수보다는 정책 요인에 의해 결정.

📌 종합 결론

옥수수: 거시 변수 전반에 양의 반응 → 비축·사료 필수재라 어떤 충격에도 수입 유지·확대.

밀: 유가·환율 중심으로 유의 → 가공식품 필수재로 수입 지속.

쌀: 거시 변수와 무관 → 정책/쿼터가 절대적.

기타 곡물: 규모 작아 환율에 민감하게 줄었다 늘었다.

```python
import pandas as pd
import statsmodels.api as sm

# 데이터 불러오기
path = "/content/master_with_IGREA_GSCPI_CPUI.csv"
df = pd.read_csv(path)

# 날짜 처리
df['date'] = pd.to_datetime(df['date'])
df = df.set_index('date')

# 분석 대상
targets = [
    "Wheat_import_weight_kg",
    "Corn_import_weight_kg",
    "Rice_import_weight_kg",
    "OtherGrains_import_weight_kg"
]

candidates = ["WTI_Oil", "Brent_Oil", "krw_per_usd",
              "USD_Index", "IGREA", "GSCPI", "CPUI"]

results = []

for target in targets:
    for var in candidates:
        # Drop NaN
        sub = df[[target, var]].dropna()

        # 회귀모형: target ~ var
        X = sm.add_constant(sub[var])  # 상수항 추가
        y = sub[target]
        model = sm.OLS(y, X).fit()

        coef = model.params[var]
        pval = model.pvalues[var]
        r2 = model.rsquared

        results.append({
            "Target": target,
            "Variable": var,
            "Coef": coef,
            "p-value": pval,
            "R2": r2
        })

# 결과 요약 테이블
results_df = pd.DataFrame(results)
print(results_df.sort_values(by="p-value").head(20))  # 유의성 높은 상위 15개
```

👉 해석: 기타 곡물은 전체적으로 물가가 오르면 수입량이 줄어드는 성향.
규모가 작고 대체재가 많아서, 물가 충격 시 먼저 줄이는 항목으로 보임.

```python
# ================== Robust loader + lead/lag heatmaps (CPI version) ==================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1) Load robustly and get 'date' ---
path = "/content/master_with_IGREA_GSCPI_CPUI.csv"  # 본인 파일 경로로 수정

def load_with_date(path):
    try:
        df = pd.read_csv(path, encoding="utf-8-sig")
    except:
        df = pd.read_csv(path)
    df.columns = [c.strip() for c in df.columns]

    if "date" in df.columns:
        df["date"] = pd.to_datetime(df["date"], errors="coerce")
        return df

    # 첫 번째 칼럼이 날짜일 가능성 처리
    df_try = pd.read_csv(path, parse_dates=[0], encoding="utf-8-sig")
    first = df_try.columns[0]
    if np.issubdtype(df_try[first].dtype, np.datetime64):
        df_try = df_try.rename(columns={first: "date"})
        return df_try

    # date 비슷한 이름 처리
    for c in df.columns:
        if "date" in c.lower():
            df = df.rename(columns={c: "date"})
            df["date"] = pd.to_datetime(df["date"], errors="coerce")
            return df

    # fallback: 2015-01부터 월별 생성
    start = pd.Timestamp("2015-01-01")
    df["date"] = pd.date_range(start=start, periods=len(df), freq="MS")
    return df

df = load_with_date(path)
df = df.dropna(subset=["date"]).sort_values("date").reset_index(drop=True)
df = df.set_index("date")

# 숫자형으로 변환
for c in df.columns:
    if not pd.api.types.is_numeric_dtype(df[c]):
        df[c] = pd.to_numeric(df[c].astype(str).str.replace(",", ""), errors="coerce")

# --- 2) Targets (grain import weights) & candidates (CPI variables) ---
targets = [
    "Wheat_import_weight_kg",
    "Corn_import_weight_kg",
    "Rice_import_weight_kg",
    "OtherGrains_import_weight_kg"
]
candidates = ["CPI_USA", "CPI_KOR", "CPI_KOR_core",
              "CPI_KOR_food_energy", "CPI_KOR_grain"]

targets = [t for t in targets if t in df.columns]
candidates = [x for x in candidates if x in df.columns]

print("Targets in file:", targets)
print("CPI variables in file:", candidates)

# --- 3) Lead/Lag correlation (±6개월) ---
max_lag = 6
def lead_lag_table(y, df_vars, xs, max_lag=6):
    res = pd.DataFrame(index=range(-max_lag, max_lag+1), columns=xs, dtype=float)
    for x in xs:
        s = df_vars[x]
        for lag in range(-max_lag, max_lag+1):
            if lag > 0:   # x leads
                xc, yc = s.shift(lag), y
            elif lag < 0: # y leads
                xc, yc = s, y.shift(-lag)
            else:
                xc, yc = s, y
            res.loc[lag, x] = xc.corr(yc)
    return res

for tgt in targets:
    tbl = lead_lag_table(df[tgt], df, candidates, max_lag=max_lag)

    # 히트맵
    plt.figure(figsize=(12,7))
    sns.heatmap(tbl.T, annot=True, fmt=".2f", cmap="coolwarm", center=0)
    plt.title(f"Lead/Lag Correlation with {tgt} vs CPI variables (±{max_lag}M)")
    plt.xlabel("Lag (months)   →   Positive = CPI leads")
    plt.ylabel("CPI variables")
    plt.tight_layout()
    plt.show()

    # Top-5 lag 결과 요약
    best = tbl.abs().stack().sort_values(ascending=False).head(5)
    print(f"[Top-5 lags | {tgt}]")
    for (lag, var), _ in best.items():
        direction = "CPI leads" if lag > 0 else "Target leads" if lag < 0 else "coincident"
        print(f"  {var:>15s} @ lag {lag:+}M  corr={tbl.loc[lag,var]:+.3f}  ({direction})")
    print()

```

4️⃣ 기타 곡물 (Other Grains)

✅ 종합 인사이트

밀·옥수수:

CPI 상승 → 수입량도 늘어나는 양(+)의 탄력적 관계 존재.

특히 옥수수는 곡물 CPI의 중심 변수.

이는 "가격 ↑ → 소비 ↓"가 아니라, 오히려 가격 ↑ → 수입(공급) 확대로 대응하는 산업적/정책적 구조 때문.

쌀:

정책적 요소가 커서 시장 물가(CPI)와 무관.

기타 곡물:

CPI 영향 미미.

```python
import pandas as pd
import statsmodels.api as sm
import numpy as np

# 1. 데이터 불러오기
path = "/content/master_with_IGREA_GSCPI_CPUI.csv"
df = pd.read_csv(path, encoding="utf-8-sig")

# 2. 날짜 처리
df['date'] = pd.to_datetime(df['date'])
df = df.set_index('date')

# 3. 분석 대상 변수
cpi_vars = ["CPI_USA", "CPI_KOR", "CPI_KOR_core",
            "CPI_KOR_food_energy", "CPI_KOR_grain"]

grain_targets = ["Wheat_import_weight_kg",
                 "Corn_import_weight_kg",
                 "Rice_import_weight_kg",
                 "OtherGrains_import_weight_kg"]

# 4. 로그 변환 (탄력성 분석을 위해)
for col in cpi_vars + grain_targets:
    df[f"log_{col}"] = np.log(df[col].replace(0, np.nan))

# 5. 단변량 회귀 함수
def run_univariate_regressions(target, predictors):
    results = []
    for var in predictors:
        data = df[[f"log_{target}", f"log_{var}"]].dropna()
        X = sm.add_constant(data[f"log_{var}"])
        y = data[f"log_{target}"]
        model = sm.OLS(y, X).fit()
        coef = model.params[1]  # log-log 계수 = 탄력성
        pval = model.pvalues[1]
        r2 = model.rsquared
        results.append([target, var, coef, pval, r2])
    return pd.DataFrame(results, columns=["Target", "Variable", "Elasticity", "p-value", "R2"])

# 6. 곡물별 회귀 실행
all_results = pd.concat([run_univariate_regressions(target, cpi_vars) for target in grain_targets])

# 7. 유의한 결과만 필터 (p<0.05)
significant_results = all_results[all_results["p-value"] < 0.05]

# 결과 출력
print("=== 전체 결과 ===")
print(all_results)

print("\n=== 통계적으로 유의한 결과만 ===")
print(significant_results)
```

###**📌VAR–IRF(충격반응) 시뮬레이션**
=>IGREA, GSCPI, CPUI CPI에 10% 쇼크가 발생했을 때, 향후 12개월 동안 곡물 수입량(kg) 이 어떻게 반응하는지 확인.

```python
# ============================================
# IRF 한눈에 보기 (IGREA/GSCPI/CPUI → 4개 곡물)
#  - 각 쇼크(IGREA, GSCPI, CPUI)별로 "한 그래프"에
#    Corn/Wheat/Rice/OtherGrains 4개 반응을 함께 표시
#  - 작은 표본에서도 돌아가도록 시차 상한(cap) 적용
#  - IRF는 기본 1σ(표준편차) 충격. 10% 충격으로 바꾸려면 SHOCK_MODE="pct"
# ============================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.api import VAR

# ===== 1) 데이터 로드 & 기본 전처리 =====
PATH = "/content/sample_data/master_with_IGREA_GSCPI_CPUI.csv"
df = pd.read_csv(PATH, encoding="utf-8-sig")

# 날짜 컬럼 잡기
date_col = [c for c in df.columns if "date" in c.lower()][0]
df[date_col] = pd.to_datetime(df[date_col], errors="coerce")
df = df.dropna(subset=[date_col]).sort_values(date_col).set_index(date_col)

# 숫자형 변환
for c in df.columns:
    if not pd.api.types.is_numeric_dtype(df[c]):
        df[c] = pd.to_numeric(df[c].astype(str).str.replace(",", ""), errors="coerce")

# 쇼크 변수 & 곡물 변수
SHOCKS = ["IGREA", "GSCPI", "CPUI"]
GRAINS = [
    "Corn_import_weight_kg",
    "Wheat_import_weight_kg",
    "Rice_import_weight_kg",
    "OtherGrains_import_weight_kg",
]
SHOCKS = [x for x in SHOCKS if x in df.columns]
GRAINS = [x for x in GRAINS if x in df.columns]

# ===== 2) 로그-차분(≈%변화) 함수 =====
def make_dlog(X):
    eps = 1e-9
    X = X.clip(lower=eps)
    return np.log(X).diff().dropna()

# ===== 3) VAR 적합 + IRF 그리기(각 쇼크별로 4개 곡물 라인) =====
MAX_LAG = 3      # 표본 작아서 4으로 함
HORIZON = 12     # 12개월
SHOCK_MODE = "stdev"  # "stdev"(기본) or "pct"
PCT = 0.10       # SHOCK_MODE="pct"일 때 사용 (+10% 충격)

def select_safe_lag(dlogX, max_lag=3):
    nobs, nvars = len(dlogX), dlogX.shape[1]
    # 너무 큰 시차 방지용 안전 상한
    safe = min(max_lag, max(1, nobs // max(4*nvars, 8)))
    if safe < 1: safe = 1
    order = VAR(dlogX).select_order(safe)
    p = order.selected_orders.get('aic', 1) or 1
    return int(min(max(1, p), safe))

def plot_irf_multi_for_shock(shock):
    cols = [shock] + GRAINS
    sub = df[cols].dropna()
    if len(sub) < 25:
        print(f"[{shock}] 표본 부족: {len(sub)}행. 스킵.")
        return

    dlogX = make_dlog(sub)
    p = select_safe_lag(dlogX, MAX_LAG)
    res = VAR(dlogX).fit(p)
    irf = res.irf(HORIZON)  # orth=True (Cholesky)

    # 스케일링: 1σ or +10%
    resid_std = pd.Series(np.sqrt(np.diag(res.sigma_u)), index=dlogX.columns)
    if SHOCK_MODE == "pct":
        scale = np.log(1+PCT) / resid_std[shock]
        title_suffix = f"(+{int(PCT*100)}% shock)"
    else:
        scale = 1.0
        title_suffix = "(1σ shock)"

    shock_idx = list(dlogX.columns).index(shock)
    arr = irf.irfs  # (h+1, nvars, nvars)

    # --- Plot ---
    plt.figure(figsize=(9,5))
    for g in GRAINS:
        ridx = list(dlogX.columns).index(g)
        path = arr[:, ridx, shock_idx] * scale          # Δlog 반응
        cum = np.cumsum(path)                           # 누적 → 수준효과
        plt.plot(range(HORIZON+1), 100*cum, label=g)    # %로 표시

    plt.title(f"IRF (Cumulative %): {shock} {title_suffix} | VAR(p={p})")
    plt.xlabel("Months after shock")
    plt.ylabel("Cumulative % change (approx.)")
    plt.axhline(0, lw=0.8, ls="--")
    plt.legend()
    plt.tight_layout()
    plt.show()

# ===== 4) 실행: IGREA / GSCPI / CPUI 각각 한 장씩 =====
for s in SHOCKS:
    plot_irf_multi_for_shock(s)

```
